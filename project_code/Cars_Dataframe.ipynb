```python
csv_path = Path("../test/car_price.csv")
cars_df = pd.read_csv(csv_path)

cars_df.head(100)
```

```python
cars_df = cars_df.set_index("car_ID")
cars_df
```

```python
cars_df = cars_df.drop(columns="symboling")
cars_df
```

```python
cars_df[['Car Make', 'Car Model']] = cars_df['CarName'].str.split(pat=" ", n=1, expand=True)
cars_df
```

```python
cars_df = cars_df.rename(columns={"Car_Make":" Car Make", "Car_Model":"Car Model","fueltype":"Fuel Type", 
                                  "aspiration": "Aspiration", "doornumber":"Number of Doors", "carbody":"Body Type", 
                                  "drivewheel":"Drivetrain", "enginelocation":"Location of Engine", "wheelbase":"Wheelbase",
                                  "carlength":"Length of Car", "carwidth":"Width of Car", "carheight":"Height of Car", 
                                  "curbweight":"Curb Weight", "enginetype":"Engine Type", "cylindernumber":"Number of Cylinders",
                                  "enginesize":"Engine Size", "fuelsystem":"Fuel System", "boreratio":"Bore Ratio", 
                                  "stroke":"Stroke", "compressionratio":"Compression Ratio", "horsepower":"Horsepower", 
                                  "peakrpm":"Peak RPM", "citympg":"City MPG", "highwaympg":"Highway MPG", "price":"Price"})

cars_df
```

```python
cars_df = cars_df.loc[:, ['CarName', "Car Make", 'Car Model','Fuel Type', 'Aspiration', 'Number of Doors', 'Body Type',
                          'Drivetrain', 'Location of Engine', 'Wheelbase', 'Length of Car',
                          'Width of Car', 'Height of Car', 'Curb Weight', 'Engine Type',
                          'Number of Cylinders', 'Engine Size', 'Fuel System', 'Bore Ratio',
                          'Stroke', 'Compression Ratio', 'Horsepower', 'Peak RPM', 'City MPG',
                          'Highway MPG', 'Price']]

cars_df
```

```python
cars_df = cars_df.drop(columns="CarName")
cars_df
```

```python
x = cars_df.loc[:,"City MPG"].values
y = cars_df.loc[:,"Highway MPG"].values

cars_df["Average MPG"] = (x + y) / 2
cars_df
```
