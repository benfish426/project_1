```python
import pandas as pd
import requests

api_key = v_api_key
base_url = 'https://mc-api.marketcheck.com/v2'

# Function to fetch cars with specified parameters
def fetch_cars_with_params(api_key, car_types=['new', 'used'], body_type='', vehicle_type='', price_range='', msrp_range='', year_range=''):
    all_cars_data = []

    for car_type in car_types:
        url = f'{base_url}/popular/cars?api_key={api_key}&car_type={car_type}&body_type={body_type}&vehicle_type={vehicle_type}&price_range={price_range}&msrp_range={msrp_range}&year_range={year_range}'
        response = requests.get(url)

        try:
            data = response.json()
            if 'popular_cars' in data:
                cars_data = data['popular_cars']
                all_cars_data.extend(cars_data)
            else:
                print(f"Error in API response for {car_type} cars: {data}")
        except Exception as e:
            print(f"Error parsing JSON response for {car_type} cars: {e}")

    return all_cars_data

# Fetch cars data with specified parameters
all_cars_data_filtered = fetch_cars_with_params(api_key, car_types=['new', 'used'], body_type='sedan', vehicle_type='car', price_range='15000-30000', msrp_range='15000-40000', year_range='2010-2022')

# Extract relevant information and create a DataFrame
all_cars_df_filtered = pd.DataFrame([
    {
        'car_type': car.get('car_type', ''),
        'year': int(car.get('year_range', '0-0').split('-')[0]),
        'make': car.get('make', ''),
        'model': car.get('model', ''),
        'body_type': car.get('body_type', ''),
        'vehicle_type': car.get('vehicle_type', ''),
        'price': car.get('price_range', {}).get('mean', 0),
        'msrp': car.get('msrp_range', {}).get('mean', 0)
    }
    for car in all_cars_data_filtered
])

# Display the DataFrame
all_cars_df_filtered.head()
```
