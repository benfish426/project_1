```python
import pandas as pd
from pathlib import Path

# Assuming the "car_details.csv" file has various columns
csv_path = Path("car_data/car_details.csv")
car_details_df = pd.read_csv(csv_path)

# Selecting specific columns
selected_columns = ['Make', 'Price', 'Location', 'Year', 'Kilometer', 'Fuel Type',]

# Creating a new DataFrame with only the selected columns
new_car_details_df = car_details_df[selected_columns]

# Display the new DataFrame
print(new_car_details_df.columns)
```

```python
import pandas as pd
from pathlib import Path

csv_path = Path("car_data/worldcities.csv")
cities_df = pd.read_csv(csv_path)

# Selecting specific columns
selected_columns = ['city', 'lat', 'lng']

# Creating a new DataFrame with only the selected columns
cities_df = cities_df[selected_columns]

# Display the new DataFrame
print(cities_df.columns)
```

```python
merged_df = pd.merge(new_car_details_df, cities_df[['city', 'lat', 'lng']], left_on='Location', right_on='city', how='left')

# Drop the redundant 'city' column
merged_df = merged_df.drop('city', axis=1)

# Display the DataFrame with longitude and latitude filled
merged_df
```

```python
column_counts = merged_df.count()

# Display the column counts
print(column_counts)
```

```python
import pandas as pd

# Assuming your merged data is stored in a DataFrame called merged_df
# You can replace 'merged_df' with the name of your DataFrame

# Filter for rows where 'lat' is missing
missing_lat_locations = merged_df[merged_df['lat'].isnull()]

# Extract unique locations from the filtered DataFrame
unique_missing_lat_locations = missing_lat_locations['Location'].unique()

# Print the locations that have missing latitude values
for location in unique_missing_lat_locations:
    print(location)
```

```python
import pandas as pd

missing_values = merged_df.isnull().sum()

# Display the missing values before deleting rows
print("Missing values before deleting rows:")
print(missing_values)

# Drop rows with missing values
df_cleaned = merged_df.dropna()
```

```python
column_counts = df_cleaned.count()

# Display the column counts
print(column_counts)
```

```python
brand_years = df_cleaned.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# Filter the DataFrame based on the specified makes
specified_makes = ['Audi', 'BMW', 'Ford', 'Honda', 'Hyundai', 'Jeep', 'Land Rover', 'MINI', 'Mercedes-Benz', 'Porsche', 'Toyota', 'Volkswagen', 'Volvo']

# Filter the selected_years_df based on specified makes
cleaned_df = df_cleaned[df_cleaned['Make'].isin(specified_makes)]

# Display the filtered DataFrame
print(cleaned_df.head())
```

```python
brand_years = cleaned_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# Filter the DataFrame based on the specified year range
selected_years_df = cleaned_df[(cleaned_df['Year'] >= 2017) & (cleaned_df['Year'] <= 2021)]

brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# List of makes to remove
makes_to_remove = ['MINI', 'Jeep', 'Volvo', 'Ford', 'Porsche']

# Filtering out rows with the specified makes
selected_years_df = selected_years_df[~selected_years_df['Make'].isin(makes_to_remove)]
brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
 clean_car_df = selected_years_df
 clean_car_df.columns
clean_car_df.head
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_filtered is the DataFrame containing the filtered data
# You may need to adjust column names according to your DataFrame

# Calculate the average price for each make and year
average_price =  clean_car_df.groupby(['Make', 'Year'])['Price'].mean().reset_index()

# Pivot the DataFrame to have years as columns and makes as rows
pivot_table = average_price.pivot(index='Make', columns='Year', values='Price')

# Plotting the bar plot
ax = pivot_table.plot(kind='bar', figsize=(12, 8))

# Formatting y-axis tick labels to display price without dollar symbol
yticks = ax.get_yticks()
yticklabels = [f'{int(ytick):,}' for ytick in yticks]
ax.set_yticklabels(yticklabels)

plt.title('Average Price by Make and Year')
plt.xlabel('Make')
plt.ylabel('Average Price in Rupees')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'clean_car_df'
# Group data by 'Year' and calculate the average price for each year
price_trends = clean_car_df.groupby('Year')['Price'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(price_trends['Year'], price_trends['Price'], marker='o', linestyle='-')
plt.title('Average Car Price Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price (Rupees)')
plt.grid(True)
plt.xticks(price_trends['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import folium

# Assuming clean_car_df is the DataFrame containing the cleaned data
# You may need to adjust column names according to your DataFrame

# Create a map centered at the mean of latitude and longitude
map_center = [clean_car_df['lat'].mean(), clean_car_df['lng'].mean()]
m = folium.Map(location=map_center, zoom_start=5)

# Add markers for each location with price as popup
for index, row in clean_car_df.iterrows():
    folium.Marker([row['lat'], row['lng']], popup=f"${row['Price']:,}").add_to(m)

# Display the map
m
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(clean_car_df['Kilometer'], clean_car_df['Price'], color='blue', alpha=0.5)
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price Rupees')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
from scipy.stats import linregress


slope, intercept, r_value, p_value, std_err = linregress(clean_car_df['Kilometer'], clean_car_df['Price'])


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Kilometer', y='Price', data=clean_car_df, color='blue', alpha=0.5)


plt.plot(clean_car_df['Kilometer'], intercept + slope * clean_car_df['Kilometer'], color='red', label='Trendline')

plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price Rupees')
plt.grid(True)
plt.legend()
plt.show()

# Print regression parameters
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value ** 2:.2f}")
print(f"P-value: {p_value:.4f}")
```
