```python
import pandas as pd
from pathlib import Path

csv_path = Path("car_data/car_details.csv")
car_details_df = pd.read_csv(csv_path)

# Display the columns of the DataFrame
print(car_details_df.columns)
```

```python
car_details_df["Location"].unique()
```

```python
car_details_df["Price"] = car_details_df["Price"]*.012
car_details_df["Kilometer"] = round(car_details_df["Kilometer"]/1.609,2)
car_details_df["Length"] = round(car_details_df["Length"]/304.8,2)
car_details_df["Width"] = round(car_details_df["Width"]/304.8,2)
car_details_df["Height"] = round(car_details_df["Height"]/304.8,2)

car_details_df = car_details_df.rename(columns={"Price":"Price USD","Kilometer":"Miles","Length":"Length (Ft)",
                                                "Width":"Width (Ft)", "Height":"Height (Ft)"})

car_details_df[['Engine Size', 'CC']] = car_details_df['Engine'].str.split(pat=" ", n=1, expand=True)

car_details_df
```

```python
car_details_df = car_details_df[['Make', 'Model', 'Price USD', 'Year', 'Miles', 'Fuel Type',
       'Transmission', 'Location', 'Color', 'Owner', 'Seller Type', 'Engine Size',
       'CC', 'Max Power', 'Max Torque', 'Drivetrain', 'Length (Ft)', 'Width (Ft)',
       'Height (Ft)', 'Seating Capacity', 'Fuel Tank Capacity']]
car_details_df
```

```python
column_counts = car_details_df.count()

# Display the column counts
print(column_counts)
```

```python
missing_values = car_details_df.isnull().sum()
# Display the missing values before deleting rows
print("Missing values before deleting rows:")
print(missing_values)
# Drop rows with missing values
df_cleaned = car_details_df.dropna()
```

```python
column_counts = df_cleaned.count()
# Display the column counts
print(column_counts)
```

```python
# Filter the DataFrame based on the specified makes
specified_makes = ['Audi', 'BMW', 'Ford', 'Honda', 'Hyundai', 'Jeep', 'Land Rover', 'MINI', 'Mercedes-Benz', 'Porsche', 'Toyota', 'Volkswagen', 'Volvo']

# Filter the selected_years_df based on specified makes
cleaned_df = df_cleaned[df_cleaned['Make'].isin(specified_makes)]

# Display the filtered DataFrame
print(cleaned_df.head())
```

```python
brand_years = cleaned_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# Filter the DataFrame based on the specified year range
selected_years_df = cleaned_df[(cleaned_df['Year'] >= 2017) & (cleaned_df['Year'] <= 2021)]

brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# List of makes to remove
makes_to_remove = ['MINI', 'Jeep', 'Volvo', 'Ford', 'Porsche']

# Filtering out rows with the specified makes
selected_years_df = selected_years_df[~selected_years_df['Make'].isin(makes_to_remove)]
brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
import seaborn as sns
from scipy.stats import linregress


slope, intercept, r_value, p_value, std_err = linregress(clean_car_df['Miles'], clean_car_df['Price USD'])


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Miles', y='Price USD', data=clean_car_df, color='blue', alpha=0.5)


plt.plot(clean_car_df['Miles'], intercept + slope * clean_car_df['Miles'], color='red', label='Trendline')

plt.title('Price vs Miles Driven')
plt.xlabel('Miles Driven')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()

# Print regression parameters
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value ** 2:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
def plot_linear_regression(x_values,y_values,title,text_coordinates):
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values,y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + " x + " + str(round(intercept,2))
    
    plt.scatter(x_values,y_values)
    plt.plot(x_values, regress_values, "r-")
    plt.annotate(line_eq, text_coordinates, color="red")
    plt.xlabel("Price USD")
    plt.ylabel(title)
    print(f"The r-squared is: {rvalue}")
    plt.show()
```

```python
df_cleaned["Price USD"] = df_cleaned["Price USD"].astype("float")
df_cleaned["Engine Size"] = df_cleaned["Engine Size"].astype("float")

x_values = df_cleaned["Price USD"]
y_values = df_cleaned["Engine Size"]

plot_linear_regression(x_values,y_values,"Engine Size",(125000,2200))
```
