# AutoValue Trends
### Analyzing Market Retention Across All Brands and Understanding Market Value Trends in Vehicles

```python
import pandas as pd
from pathlib import Path

csv_path = Path("car_data/car_details.csv")
car_details_df = pd.read_csv(csv_path)

# Display the columns of the DataFrame
print(car_details_df.columns)
```

```python
car_details_df
```

```python
car_details_df["Location"].unique()
```

```python
car_details_df["Price"] = car_details_df["Price"]*.012
car_details_df["Kilometer"] = round(car_details_df["Kilometer"]/1.609,2)
car_details_df["Length"] = round(car_details_df["Length"]/304.8,2)
car_details_df["Width"] = round(car_details_df["Width"]/304.8,2)
car_details_df["Height"] = round(car_details_df["Height"]/304.8,2)

car_details_df = car_details_df.rename(columns={"Price":"Price USD","Kilometer":"Miles","Length":"Length (Ft)",
                                                "Width":"Width (Ft)", "Height":"Height (Ft)"})

car_details_df
```

```python
column_counts = car_details_df.count()

# Display the column counts
print(column_counts)
```

```python
missing_values = car_details_df.isnull().sum()
# Display the missing values before deleting rows
print("Missing values before deleting rows:")
print(missing_values)
# Drop rows with missing values
df_cleaned = car_details_df.dropna()
```

```python
column_counts = df_cleaned.count()
# Display the column counts
print(column_counts)
```

```python
df_cleaned
```

```python
column_counts = df_cleaned.count()

# Display the column counts
print(column_counts)
```

```python
brand_years = df_cleaned.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# Filter the DataFrame based on the specified makes
specified_makes = ['Audi', 'BMW', 'Ford', 'Honda', 'Hyundai', 'Jeep', 'Land Rover', 'MINI', 'Mercedes-Benz', 'Porsche', 'Toyota', 'Volkswagen', 'Volvo']

# Filter the selected_years_df based on specified makes
cleaned_df = df_cleaned[df_cleaned['Make'].isin(specified_makes)]

# Display the filtered DataFrame
print(cleaned_df.head())
```

```python
brand_years = cleaned_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# Filter the DataFrame based on the specified year range
selected_years_df = cleaned_df[(cleaned_df['Year'] >= 2017) & (cleaned_df['Year'] <= 2021)]

brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
# List of makes to remove
makes_to_remove = ['MINI', 'Jeep', 'Volvo', 'Ford', 'Porsche']

# Filtering out rows with the specified makes
selected_years_df = selected_years_df[~selected_years_df['Make'].isin(makes_to_remove)]
brand_years = selected_years_df.groupby('Make')['Year'].unique()

# Displaying the result
print(brand_years)
```

```python
clean_car_df = selected_years_df
clean_car_df.columns
clean_car_df.head
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_filtered is the DataFrame containing the filtered data
# You may need to adjust column names according to your DataFrame

# Calculate the average price for each make and year
average_price =  clean_car_df.groupby(['Make', 'Year'])['Price USD'].mean().reset_index()

# Pivot the DataFrame to have years as columns and makes as rows
pivot_table = average_price.pivot(index='Make', columns='Year', values='Price USD')

# Plotting the bar plot
ax = pivot_table.plot(kind='bar', figsize=(12, 8))

# Formatting y-axis tick labels to display price without dollar symbol
yticks = ax.get_yticks()
yticklabels = [f'{int(ytick):,}' for ytick in yticks]
ax.set_yticklabels(yticklabels)

plt.title('Average Price by Make and Year')
plt.xlabel('Make')
plt.ylabel('Average Price in USD')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'clean_car_df'
# Group data by 'Year' and calculate the average price for each year
price_trends = clean_car_df.groupby('Year')['Price USD'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(price_trends['Year'], price_trends['Price USD'], marker='o', linestyle='-')
plt.title('Average Car Price Over Time')
plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.grid(True)
plt.xticks(price_trends['Year'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(clean_car_df['Miles'], clean_car_df['Price USD'], color='blue', alpha=0.5)
plt.title('Price vs Miles Driven')
plt.xlabel('Miles Driven')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
from scipy.stats import linregress


slope, intercept, r_value, p_value, std_err = linregress(clean_car_df['Miles'], clean_car_df['Price USD'])


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Miles', y='Price USD', data=clean_car_df, color='blue', alpha=0.5)


plt.plot(clean_car_df['Miles'], intercept + slope * clean_car_df['Miles'], color='red', label='Trendline')

plt.title('Price vs Miles Driven')
plt.xlabel('Miles Driven')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()

# Print regression parameters
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value ** 2:.2f}")
print(f"P-value: {p_value:.4f}")
```
